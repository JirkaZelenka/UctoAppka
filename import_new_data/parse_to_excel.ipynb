# Text / Markdown → Excel

Nastav `VSTUP` a `START_ROK`, spusť buňky. Logika v `parse_md_txt_to_excel.py` (u dat bez roku se rok doplňuje z chronologie).

```python
from pathlib import Path
import pandas as pd
from parse_md_txt_to_excel import (
    parse_file,
    parse_denni_file,
    DEFAULT_START_YEAR,
    SLOUPCE,
    SLOUPCE_DENNI,
)

VSTUP = Path("all_data.txt")
VYSTUP = VSTUP.with_suffix(".xlsx")
# první datumová hlavička bez roku: rok START_ROK; další se odvíjí od chronologie
START_ROK = DEFAULT_START_YEAR
```

```python
rows = parse_file(VSTUP, start_rok=START_ROK)
df = pd.DataFrame(rows, columns=list(SLOUPCE))
if len(df) and "datum" in df.columns:
    df["datum"] = pd.to_datetime(df["datum"], errors="coerce").dt.date
df.to_excel(VYSTUP, index=False, engine="openpyxl")
print(f"Zapsáno: {VYSTUP.resolve()} ({len(df)} řádků)")
df
```

```python
VYSTUP_DENNI = VYSTUP.parent / f"{VYSTUP.stem}_denni{VYSTUP.suffix}"
rows_d = parse_denni_file(VSTUP, start_rok=START_ROK)
df_d = pd.DataFrame(rows_d, columns=list(SLOUPCE_DENNI))
if len(df_d) and "datum" in df_d.columns:
    df_d["datum"] = pd.to_datetime(df_d["datum"], errors="coerce").dt.date
df_d.to_excel(VYSTUP_DENNI, index=False, engine="openpyxl")
print(f"Zapsáno (denní): {VYSTUP_DENNI.resolve()} ({len(df_d)} řádků)")
df_d
```
